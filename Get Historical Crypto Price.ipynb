```python
import pandas as pd
from datetime import datetime as dt
import time
import requests
import json
```

```python
cryptoMarketCap = pd.read_csv('data/CryptoMarketCap.csv')
CryptoList = list(set(cryptoMarketCap[cryptoMarketCap['Rank'] <= 50]['Symbol']))
```

```python
# https://docs.cryptowat.ch/rest-api/markets/ohlc
headers = {
    'X-CW-API-Key': 'GY5RZR8WXHXSDHEQ9395'
}

x = requests.get("https://api.cryptowat.ch/exchanges", headers=headers)
pd.DataFrame(json.loads(x.content)['result'])
```

```python
startDate = dt(2017, 12, 31, 12, 00)
params = {
    'after': int(time.mktime(startDate.timetuple())),
    'periods': 86400
}
```

```python
result = pd.DataFrame()
for crypto in CryptoList:
    time.sleep(0.5)
    print(crypto)
    try:
        JSONdata = requests.get("https://api.cryptowat.ch/markets/coinbase/" + crypto + "USD/ohlc", params=params, headers=headers)
        dictData = json.loads(JSONdata.content)['result']['86400']
        df = pd.DataFrame(dictData, columns = ['CloseTime', 'OpenPrice', 'HighPrice', 'LowPrice', 'ClosePrice', 'Volume', 'QuoteVolume'])
        df['CloseTime'] = pd.to_datetime(df['CloseTime'], unit='s')
        df[crypto + 'USD'] = df['ClosePrice']
        df = df.set_index('CloseTime')
        result = pd.merge(df[crypto + 'USD'], result, how='outer', left_index=True, right_index=True)
    except:
        print(crypto + '----Error')
```

```python
result.to_csv('data/CoinbaseCryptoPrice.csv')
```
