```python
import pandas as pd
import requests
import json
import time
```

```python
categoriesData = requests.get('https://api.coingecko.com/api/v3/coins/categories/list')
categoriesList = pd.DataFrame(json.loads(categoriesData.content))
```

```python
result = pd.DataFrame()
for idx in categoriesList.index:
    time.sleep(5.5)
    print(categoriesList['name'][idx])
    params = {
    "vs_currency": 'usd',
    'category': categoriesList['category_id'][idx]
    }
    x = requests.get('https://api.coingecko.com/api/v3/coins/markets', params=params)
    df = pd.DataFrame(json.loads(x.content))['symbol']
    df.name = categoriesList['name'][idx]
    result = pd.concat([result, df], axis=1)
```
